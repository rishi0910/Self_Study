```python
# hangman 
print('Welcome to the Hangman game ')
print('Choose a word and guess the letters ')

word = 'p y t h o n'
target = '_ _ t h _ n'
alphabet = {}.fromkeys('abcdefghijklmnopqrstuvwxyz', False)
print(target)
att = 6
while att > 0:
    print(f"Guess the missing letters, attempts left {att}")
    guess = input()
    if guess == 'p':
        alphabet['p'] = True
        if alphabet['y'] and alphabet['o']:
            print('p y t h o n')
            print(f'You Won, with attempts {att}')
            break
        elif alphabet['y']:
            print('p y t h _ n')
            
        elif alphabet['o']:
            print('p _ t h o n')
        else:
            print('p _ t h _ n') 
    elif guess == 'y':
        alphabet['y'] = True
        if alphabet['o'] and alphabet['p']:
            print('p y t h o n')
            print(f'You Won, with attempts {att}')
            break
        elif alphabet['p']:
            print('p y t h _ n')
            
        elif alphabet['o']:
            print('_ y t h o n')
        else:
            print('_ y t h _ n')

    elif guess == 'o':
        alphabet['o'] = True
        if alphabet['p'] and alphabet['y']:
            print('p y t h o n')
            print(f'You Won, with attempts left {att}')
            break
        elif alphabet['p']:
            print('p _ t h o n')
            
        elif alphabet['y']:
            print('_ y t h o n')
        else:
            print('_ _ t h o n')

    att -= 1        

else:
    print('You lost, Game over ')
```
