```python
# nested loop 
def table(n):
    for i in range(1, n+1):
        for j in range(1, 11):
            print(f'{i} x {j} = {i * j}', end= '\t')
        print()
# main code 
val = 6
table(val)
```

```python
def table(n, i=1, j=1):
    if j == 11 or i >= n+1:
        return
    print(f'{i} x {j} = {i * j}', end= '\t')
    j += 1
  
    table(n, i, j)
    j = 1
    print()
    i +=  1
    table(n, i, j=1)

    



# main code 
val = 4
table(val)
```

```python
from abc import ABC, abstractmethod 

class Bird(ABC): 

    def __init__(self, name): 

        self.name = name 

    @abstractmethod 

    def fly(self): 

        pass 

class Sparrow(Bird): 

    def fly(self): 

        return "Sparrow is flying" 

class Penguin(Bird): 

    pass 

bird1 = Sparrow("Jack") 

bird2 = Penguin("Pengy") 

print(bird1.fly()) 

print(bird2.fly()) 
```
